```python
# Task 2: Write a program that takes an integer input from the user. Use a for loop to print all numbers from that integer down to 0.
# 1) ask the user for a number
# 2) use a for loop that iterates through the range of the number untill the iterator reaches zero
# ------------------------------------------------------------------ #

num = int(input("Enter any integer number: "))
# note : range(start, stop, step)
for i in range(num, -1, -1): print(i)
```

```python
# Task 3: Write a program that takes an integer input from the user. Use a while loop to keep doubling the number until it is greater than 1000, then print the result.
 # 1) use a while loop --> condition controlled --> keep going until the number is greater than 1000
 # 2) stop the loop the print the result
 # 3) the body of the loop will double the number ----> multiply by 2
 # ---------------------------------------------------------------- #
x = int(input("Enter an integer number: "))
while x <= 1000 :
  x *= 2
print(x)
```

```python
# Task 8: Write a program that uses a for loop to take an integer input from the user and print the factorial of that number.
# what is a factorial ?
# the factoriaal of 5 is 5 * 4 * 3 * 2 * 111
# so the faaactoril of n is n * (n - 1) * (n - 2) ....* 1
x = int(input("enter an integer number: "))
fact_x = 1
if x == 0:
  print(fact_x)
else:
  for i in range(x, 1, -1):
    fact_x *= i
  print(fact_x)







```

```python
# Task 9: Write a program that takes a number as input from the user. If the number is positive, print "Positive". If the number is negative, print "Negative". If the number is zero, print "Zero".
def num_kind(num):
  if num > 0:
    print("Positive!")
  elif num < 0 :
    print("Negative!")
  elif num == 0:
    print("Zero!")

print(num_kind(0))
```
